# First look on data

```python
import numpy as np
from matplotlib import pyplot as plt
from utils.helper import read_data
import utils.processing as processing
import matplotlib.pyplot as plt
```

```python
FOLDER = 'data/'
data_empty = read_data(FOLDER + "empty" + ".csv")
data_1p = read_data(FOLDER + "1p" +  ".csv")
data_2p = read_data(FOLDER + "2p" + ".csv")
data_3p = read_data(FOLDER + "3p" + ".csv")
```

```python
print(f"Shape empty: {data_empty.shape}")
print(f"Shape 1 person: {data_1p.shape}")
print(f"Shape 2 person: {data_2p.shape}")
```

## Nobody

```python
fig, axs = plt.subplots(1, 5,figsize=(15,10),sharex=True, sharey=  True)
# fig.suptitle('Range-Doppler Plot')
sample = data_empty[20].squeeze()
sample = np.sum(sample, axis = 1).squeeze()



for i in range(1):
    for j in range(5):
        to_plot = np.abs(sample)[j,:,:]
        pos = axs[j].imshow(to_plot)
        fig.colorbar(pos, ax=axs[j])
        axs[j].set_aspect('equal')

plt.subplots_adjust(hspace=0)
```
