## Import Libraries

```python
%matplotlib inline
import numpy as np
from utils.helper import read_data, blur, to_real, morphology
import matplotlib.pyplot as plt
FOLDER = "data/"
NUMBER_OF_RADARS = 3
```

## Load data

```python
data_empty = read_data(FOLDER + "empty.csv")
data_1p = read_data(FOLDER + "1p.csv")
print("----------------Shapes----------------")
print("empty:", data_empty.shape)
print("1p:", data_1p.shape)
```

```python
LENGTH_EMPTY = data_empty.shape[0]
LENGTH_1P = data_1p.shape[0]
```

```python
data_empty_summed = to_real(np.sum(data_empty, axis=1)) 
data_1p_summed = to_real(np.sum(data_1p, axis=1))
```

```python
data_empty_summed[:, :, 32, :] = 0
```

### Blurring

```python
data_empty_blurred = blur(data_empty_summed)
data_1p_blurred = blur(data_1p_summed)
data_empty_morph = morphology(data_empty_summed)
data_1p_morph = morphology(data_1p_summed)
```

### Plot some blurring examples

```python
s = (10,1)
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,10))
axs[0].imshow(data_empty_summed[s])
axs[0].set_title("empty")
axs[1].imshow(data_empty_blurred[s])
axs[1].set_title("empty blurred")
axs[2].imshow(data_empty_morph[s])
axs[2].set_title("empty morph")
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,10))
axs[0].imshow(data_1p_summed[s])
axs[0].set_title("1p")
axs[1].imshow(data_1p_blurred[s])
axs[1].set_title("1p blurred")
axs[2].imshow(data_1p_morph[s])
axs[2].set_title("1p morph")
```
