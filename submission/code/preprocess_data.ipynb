## Import Libraries

```python
%matplotlib inline
import numpy as np
from utils.helper import read_data, blur, morphology
import matplotlib.pyplot as plt
FOLDER = "data/"
NUMBER_OF_RADARS = 3
```

## Load data

```python
data_empty = read_data(FOLDER + "empty.csv")
data_1p = read_data(FOLDER + "1p.csv")
data_2p = read_data(FOLDER + "2p.csv")
data_3p = read_data(FOLDER + "3p.csv")
print("----------------Shapes----------------")
print("empty:", data_empty.shape)
print("1p:", data_1p.shape)
print("2p:", data_2p.shape)
print("3p:", data_3p.shape)
```

```python
LENGTH_EMPTY = data_empty.shape[0]
LENGTH_1P = data_1p.shape[0]
LENGTH_2P = data_2p.shape[0]
LENGTH_3P = data_3p.shape[0]
```

```python
data_empty_summed = np.abs(np.sum(data_empty, axis=1)) 
data_1p_summed = np.abs(np.sum(data_1p, axis=1))
data_2p_summed = np.abs(np.sum(data_2p, axis=1)) 
data_3p_summed = np.abs(np.sum(data_3p, axis=1))
```

### Blurring

```python
data_empty_blurred = blur(data_empty_summed)
data_1p_blurred = blur(data_1p_summed)
data_2p_blurred = blur(data_2p_summed)
data_3p_blurred = blur(data_3p_summed)

data_empty_morph = morphology(data_empty_summed)
data_1p_morph = morphology(data_1p_summed)
data_2p_morph = morphology(data_2p_summed)
data_3p_morph = morphology(data_3p_summed)
```

### Plot some blurring examples

```python
s = (10,1)
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,10))
axs[0].imshow(data_empty_summed[s])
axs[0].set_title("empty")
axs[1].imshow(data_empty_blurred[s])
axs[1].set_title("empty blurred")
axs[2].imshow(data_empty_morph[s])
axs[2].set_title("empty morph")
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,10))
axs[0].imshow(data_1p_summed[s])
axs[0].set_title("1p")
axs[1].imshow(data_1p_blurred[s])
axs[1].set_title("1p blurred")
axs[2].imshow(data_1p_morph[s])
axs[2].set_title("1p morph")
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,10))
axs[0].imshow(data_1p_summed[s])
axs[0].set_title("2p")
axs[1].imshow(data_1p_blurred[s])
axs[1].set_title("2p blurred")
axs[2].imshow(data_1p_morph[s])
axs[2].set_title("2p morph")
```

```python
fig, axs = plt.subplots(1,3, figsize=(15,10))
axs[0].imshow(data_1p_summed[s])
axs[0].set_title("3p")
axs[1].imshow(data_1p_blurred[s])
axs[1].set_title("3p blurred")
axs[2].imshow(data_1p_morph[s])
axs[2].set_title("3p morph")
```
